# 정규분포란?

* 정규분포(standard distribution)는 통계학에서 대표적인 연속 확률분포이며, 매우 널리 사용된다. 

* **가우스분포(The Gaussian Distribution)**라고 불리운다.



+ 정규분포의 확률밀도함수
$$ 확률밀도함수: f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\;exp\;(-\frac{(x-\mu)^2}{2\sigma^2}), \qquad   -\infty < x < \infty \qquad(\mu는 평균, \;\sigma는 \;표준편차이다.)$$

$$\int_{-\infty}^\infty f(x) \; dx = 1$$
- 확률밀도함수이므로 적분하면 1이 됨.

* 표준정규분포에서는 평균이 0, 표준편차가 1이다.

## 정규분포의 모양

image.png

```python
import numpy as np
```

```python
randn_number = np.random.randn(100) 
randn_number
```

```python
import matplotlib.pyplot as plt

plt.hist(randn_number)
plt.show()
```

* 표준정규분포를 따르는 난수를 생성하기 때문에

    가운데가 0인 모습, 정규분포는 평균을 중심으로 좌우대칭이다.

---

```python
randn_number = np.random.randn(1000)
randn_number
```

```python
import matplotlib.pyplot as plt

plt.hist(randn_number)
plt.show()
```

* 샘플수가 많아지니 점점더 정규분포의 형태로 닮아가는 모습

    마찬가지로 평균을 대칭으로 생긴 모양을 보이고 있다.

결론
---

* `randn`은 가우시안 분포를 **따르는 숫자**들 중 random으로 뽑아내는 것이다.

* 가우시안 분포를 따르는 숫자니까, 가우시안 분포 확률밀도함수 내에서 숫자를 랜덤하게 추출하게 된다.  

    따라서 범위는 ***마이너스 무한대 부터 플러스 무한대***가 된다.

* 범위는 마이너스 무한대 부터 플러스 무한대이지만, 어떠한 값(`x`)이 그 범위에 있을 확률은 

    극히 적기 때문에 눈으로 보이지가 않는 것이다.

```python
# 사용자가 평균과 표준편차를 정의할 수 있음
# 평균이 500이고 표준편차가 5인 정규분포에서 추출된 1000개의 난수 값으로 구성된 벡터

a = 5
b = 500

randn_number2 = a * np.random.randn(1000) + b
```
