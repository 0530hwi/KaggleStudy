# 제1장 자료의 정리방법

## (1-1) 서론

📌 **통계학**은 단순한 자료의 정리만을 뜻하는 것은 아니다.

예를 들어, 수천 명의 표본조사를 통하여 가구당 월수입을 조사하기 위해서는<br> **어떻게 표본설계를 하여야** 평균수입의 참값을 보다 정확히 **추정**할 수 있는가를 연구한다.

> 현대통계학에서 핵심이 되는 분야<br>
1. 추측통계학(inferential statistics): 자료에 내포되어 있는 정보를 분석하여 불확실한 사실에 대한 추론을 하는 분야<br>
(통계적 모형 설정, 설정된 모형이 합리적인지 여부 평가, 미지의 특성에 대한 결론 짓기, 미래에 일어날 현상에 대한 예측)<br><br>
2. 기술통계학(descriptive statistics): 자료를 수집하고 정리하여 도표나 표를 만들고, <br>자료를 요약하여 대표값이나 변동의 크기 등을 구하는 방법

> 자료의 요약과 표현<br>
1. 도표나 그래프를 만든다.
2. 그래프화한 자료의 대칭성을 비롯한 전반적인 형태를 고찰한다.
3. 그래프에서 일반적인 측정값과 다른 이상점에 대해 고찰한다.
4. 대표값이나 산포도와 같은 수치적 측도를 계산해본다.

## (1-2) 모집단과 표본

* 모집단(population): 연구대상이 되는 모든 가능한 관측값이나 측정값의 집합
* 표본(sample): 모집단에서 실제로 추출한 관측값이나 측정값의 집합
* 추출단위(sampling unit): 모집단의 가능한 관측값이나 측정값이 얻어지는 개체<br>
예를 들어 어느 공정에서 생산되는 전구의 수명을 검사하고자 한다.
    - 이 때 하나 하나의 전구가 **추출단위**, <br>검사를 받기 위해 추출된 일부 전구가 **표본**, <br>공정에서 생산되는 모든 전구가 **모집단**
* 유한모집단(finite population): 모집단이 유한개의 추출단위를 가질 때의 경우
* 무한모집단(infinite population): 모집단이 무한개의 추출단위를 가질 때의 경우

**✔ 주의해야 할 점**<br>
통계적 모집단은 꼭 실존하는 개체들의 집합인 것은 아니다.<br>
예를 들어 작물 재배의 경우, 지구상의 모든 지역에서 특수한 기후 조건하에 <br>매년 반복하여 경작했을 때의 수확량이 실제로 존재하는 것은 아니다. 이럴 경우 모집단은 추상적인 개념이 된다!<br>
그러나, 표본은 주어진 기후 조건하에서 소수의 지역에 경작하여 얻은 수확량을 측정하여 얻은 것으로 무한 모집단의 일부가 된다.

## (1-5) 중심위치 산포의 측도

측정형 자료의 경우, 도수분포표로부터 자료의 특성을 파악할 수도 있지만 측정값들을 객관적으로 대표할 수 있는 측도(measure)들을 많이 사용한다.<br>
흔히 대표값이라고 불리운다.

대표값에는 평균(mean), 중앙값(median), 기하평균(geometric mean) 등이 있다.

> 중요❗<br>
평균은 이상점에 의해 크게 영향을 받지만, 중앙값은 별로 영향을 받지 않는다.<br>
따라서 분포상태가 극도로 비대칭인 경우에는 중앙값이 평균보다 대표값으로서 더 큰 의미를 갖는다.<br>
그러나 중앙값으로는 모집단을 추론하기가 어렵기 때문에 어느정도 대칭이 되면 평균을 사용하는 것이 좋다.<br><br>
중앙값과 같이 평균의 이상점에 대한 예민성을 감소시키기 위해 제안된 위치축도로서 **절사평균(trimmed mean)**을 들 수 있다.

**절사평균**: $0 \leq \alpha \leq 0.05$일 때 관측값의 양쪽에서 일정비율 $\alpha$의 극단값(바깥쪽의 간측값)을 버린 <br>나머지 관측값들의 평균을 $100\alpha\%$ 절사평균이라고 한다.<br>

[절사평균](https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=lucifer246&logNo=206137587)

사분위수와 백분위수:<br>

제1사분위 $Q_1$ = 제25백분위수
제2사분위 $Q_2$ = 제50백분위수
제3사분위 $Q_3$ = 제75백분위수<br>

제 $P$백분위수라 함은 자료를 크기 순서로 늘어 놓았을 때 적어도 $P\%$의 관측값이 그 값보다 작거나 같고,<br> 또한 적어도 $(100-P)\%$의 관측값이 그 값보다 크거나 같게 되는 값을 말한다.<br>

자료의 범위와 사분위범위:<br>
(a) 범위 = 최대값 - 최소값<br>
(b) 사분위범위(IQR) = $Q_3 - Q_1$

평균 이외에 대표값이 같더라도 자료의 분표형태는 아주 다를 수 있다.<BR>
따라서 개개의 자료가 중심위치에서 얼마만큼 떨어져 있는가를 알아 볼 필요가 있다.<br>
편차의 합은 0이므로 $n$개의 편차 중에서 $(n-1)$개의 편차를 안다면 나머지 1개의 편차는 저절로 결정된다.<br>
이것은 $n$개의 값 중에서 오직 $n-1$개만이 무관함을 보여준다. 따라서 이들의 제곱의 합을 $n-1$로 나누어주어<br>
$s^2=\sum(x_i - \overline{x})^2 / (n-1)$을 자료의 **분산**이라고 정의한다.
